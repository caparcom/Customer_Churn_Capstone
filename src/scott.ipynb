## Function to handle NCES error codes

```python
# function to clean NCES error codes
def clean_nces_error_codes(main_df, cols):
    error_values = [
    "M", "N", "-1", "-2", "-9",
    "Missing", "Not applicable", "Not Applicable",
    -1, -2, -9
]
    main_df[cols] = main_df[cols].replace(error_values, np.nan)
    return main_df

```

```python
# clean ALL numeric columns (float + int)
numeric_cols = float_cols + int_cols
main_df = clean_nces_error_codes(main_df, numeric_cols)
```

```python
# convert ints safely
for col in int_cols:
    main_df[col] = pd.to_numeric(main_df[col], errors="coerce").astype("Int64")
```

## Adding what Caleb had

Operational Filtering - Only keep schools operational in all years

```python
def mark_always_operational(df):
    total_years = df["SURVYEAR"].nunique()
    
    operational_counts = (
        df[df["SY_STATUS_TEXT"] == "Currently operational"]
        .groupby("NCESSCH")["SURVYEAR"]
        .nunique()
    )
    always_operational = operational_counts[
        operational_counts == total_years
    ].index
    df["concurrently_operational"] = df["NCESSCH"].isin(always_operational)
    return df
```

```python
main_df = mark_always_operational(main_df)

main_df = main_df[main_df["concurrently_operational"]]
```

Remove Charter & Magnet Schools

```python
main_df = main_df[
    main_df["CHARTER_TEXT"].isin(["No", np.nan])
]

main_df = main_df[
    main_df["MAGNET_TEXT"].isin(["No", np.nan])
]
```

```python
main_df.drop(columns=["CHARTER_TEXT", "MAGNET_TEXT"], inplace=True)
```

Drop Redundant Columns

```python
redundant_cols = [
    "X", "Y", "OBJECTID", "ST_LEAID",
    "LEA_NAME", "LSTREET1", "LSTREET2", "LSTREET3",
    "LZIP4", "PHONE",
    "AMALM","AMALF","ASALM","ASALF",
    "BLALM","BLALF","HPALM","HPALF",
    "HIALM","HIALF","TRALM","TRALF",
    "WHALM","WHALF"
]

main_df.drop(columns=redundant_cols, inplace=True)
```

## Convert start year to numeric and sort by school

```python
# extract start Year - convert to int for sorting
main_df['year_start'] = main_df['SURVYEAR'].str[:4].astype(int)
```

```python
# sort by school and year
main_df = main_df.sort_values(['NCESSCH','year_start'])
```

## Convert all numeric columns

```python
# create list of numeric columns for conversion
numeric_cols = [
    # geographic
    'X','Y','LATCOD','LONCOD',

    # enrollment totals
    'TOTAL','TOTMENROL','TOTFENROL','MEMBER',

    # staffing
    'FTE','STUTERATIO',

    # socioeconomic
    'TOTFRL','FRELCH','REDLCH','DIRECTCERT',

    # grade levels
    'PK','KG','G01','G02','G03','G04','G05',
    'G06','G07','G08','G09','G10','G11','G12','G13',
    'UG','AE',

    # race totals
    'AM','AS','BL','HP','HI','TR','WH',

    # race gender splits
    'AMALM','AMALF',
    'ASALM','ASALF',
    'BLALM','BLALF',
    'HPALM','HPALF',
    'HIALM','HIALF',
    'TRALM','TRALF',
    'WHALM','WHALF',

    # lowest/highest grade level the school serves
    'GSLO','GSHI'
]
```

```python
# convert all numeric columns to numeric, coercing errors to NaN
main_df[numeric_cols] = main_df[numeric_cols].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
# confirm all numeric columns are now numeric
main_df[numeric_cols].dtypes.value_counts()
```

```python
# NaN for all numberic fields is 0, which means all non-numeric values were successfully converted to NaN, and there are no non-numeric values that were converted to 0.
main_df[numeric_cols].isna().sum().sort_values(ascending=True).head(15)
```

```python
main_df[numeric_cols].isna().sum().sort_values(ascending=False).head(15)
```

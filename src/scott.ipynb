## Convert all numeric columns

```python
# To Do: once done inspecting turn the following to int as they dont matter for inspection and will be easier to work with as numeric types
# ObjectID: Internal GIS Identifier (int)
# GSLO  Lowest grade served (int)
# GSHI  Highest grade served (int)
```

```python
# create list of numeric columns for conversion
numeric_cols = [
    
    # enrollment totals 
    'TOTAL','TOTMENROL','TOTFENROL','MEMBER',

    # socioeconomic
    'TOTFRL','FRELCH','REDLCH','DIRECTCERT',

    # grade levels
    'PK','KG','G01','G02','G03','G04','G05',
    'G06','G07','G08','G09','G10','G11','G12','G13',
    'UG','AE',

    # race totals
    'AM','AS','BL','HP','HI','TR','WH',

    # race gender splits
    'AMALM','AMALF',
    'ASALM','ASALF',
    'BLALM','BLALF',
    'HPALM','HPALF',
    'HIALM','HIALF',
    'TRALM','TRALF',
    'WHALM','WHALF',
]
```

```python
# convert all numeric columns to numeric, coercing errors to NaN
main_df[numeric_cols] = main_df[numeric_cols].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
# confirm all numeric columns are now numeric
main_df[numeric_cols].dtypes.value_counts()
```

## Now that we have everything to numeric, change -2 to NaN so we can do checks and balances

```python
# what columns have -2 values, which is the code for "data not available" in our data dictionary
cols_with_neg2 = [
    col for col in numeric_cols
    if (main_df[col] == -2).any()
]

cols_with_neg2
```

```python
# how many -2 values are in each column that has them?
for col in cols_with_neg2:
    print(col, (main_df[col] == -2).sum())
```

```python
# replace placeholder -2 values with NaN for all numeric columns
main_df[numeric_cols] = main_df[numeric_cols].replace(-2, np.nan)
```

```python
# verify that -2 values have been converted to NaN
for col in cols_with_neg2:
    print(col, (main_df[col] == -2).sum())
```

```python
# now that -2 values have been replaced with NaN, we can check which columns have the most missing values
main_df[numeric_cols].isna().sum().sort_values(ascending=False).head(10)
```

## Grade Totals

    Sum of grade_cols  TOTAL

```python
# grade totals vs total enrollment
grade_cols = [
    'PK','KG','G01','G02','G03','G04','G05',
    'G06','G07','G08','G09','G10','G11','G12','G13',
    'UG','AE'
]

(main_df[grade_cols].sum(axis=1) - main_df['TOTAL']).describe()
```

```python
# how many are zero and below zero
(main_df[grade_cols].sum(axis=1) - main_df['TOTAL']).value_counts().head(20)
```

```python
# how many rows are below zero
diff = main_df[grade_cols].sum(axis=1) - main_df['TOTAL']
(diff != 0).sum()
```

```python
# check percentage of how many rows are below zero
(diff != 0).mean()
```

```python
# storing: how many are zero and below zero
main_df['grade_total_diff'] = (
    main_df[grade_cols].sum(axis=1) - main_df['TOTAL']
)
```

```python
# create a df of rows that grade totals do not match total enrollment
problem_rows = main_df[main_df['grade_total_diff'] != 0]
problem_rows.shape
```

```python
# sort mismatched rows by severity
problem_rows.sort_values('grade_total_diff').head(10)
```

```python
# view the most severe mismatch
problem_rows.sort_values('grade_total_diff').iloc[0]
```

```python
# view the most severe mismatch
problem_rows.sort_values('grade_total_diff').iloc[0][
    ['TOTAL'] + grade_cols
]
```
